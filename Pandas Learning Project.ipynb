# Merge, Join, and Concatenate

```python
df1 = pd.read_csv(r"C:\Users\kevin\Downloads\LOTR.csv")
df2 = pd.read_csv(r"C:\Users\kevin\Downloads\LOTR 2.csv")
```

```python
df1
```

```python
df2
```

```python
df1.merge(df2) #inner join - same data in both - default
```

```python
df1.merge(df2, how='inner', on='FellowshipID') #seprated into two spearte columns
```

```python
df1.merge(df2, how='inner', on = ['FellowshipID','FirstName']) # like the default
```

```python
df1.merge(df2, how='inner') # Inner join
```

```python
df1.merge(df2, how='outer') # outer join - all values regarless if they are same

#Nan: Not a Number
```

```python
df1.merge(df2, how='left') 
# left outer join - all values from left data frame and the overlapping data from both data frames
```

```python
df1.merge(df2, how='right') 
# right outer join - all values from right data frame and the overlapping data from both data frames
```

```python
df1.merge(df2, how='cross') 
# compares each values from both data frames, takes first row in df1 and then it compares it to every row in df2 and then
# keeps repeating untill all of the rows have been compared 
```

```python
# df1.join(df2) - will not work since it has no defaults 
#ValueError: columns overlap but no suffix specified: Index(['FellowshipID', 'FirstName'], dtype='object')
```

```python
df1.join(df2, on='FellowshipID', how = 'outer', lsuffix='_Left', rsuffix='_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix='_Left', rsuffix='_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2]) # put one ontop of each other
```

```python
pd.concat([df1,df2], join = 'inner')
```

```python
pd.concat([df1,df2], join = 'outer', axis =1)
```

# Data Cleaning

```python
df = pd.read_excel(r"C:\Users\kevin\Downloads\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns='Not_Useful_Column')
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.strip("...")
#df["Last_Name"] = df["Last_Name"].str.strip("/")
#df["Last_Name"] = df["Last_Name"].str.strip("_")
df["Last_Name"] = df["Last_Name"].str.strip("123._/")

df
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')

# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',' , 2, expand = True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace("Yes", "Y")
df['Paying Customer'] = df['Paying Customer'].str.replace("No", "N")
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace("Yes", "Y")
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace("No", "N")

df
```

```python
df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace = True)
df
```

```python
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)
df
```

```python
df = df.reset_index()
df
```

```python
df = df.reset_index(drop=True)
df
```
